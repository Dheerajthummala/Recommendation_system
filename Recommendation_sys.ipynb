```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('book.csv',encoding='Latin1')
data
```

```python
data=data.iloc[:,1:]
data
```

```python
# sorting the data
data.sort_values(by ='User.ID').reset_index(drop = True)
data
```

```python
# number of unique users in the dataset
len(data['User.ID'].unique())
```

```python
# number of unique books in the dataset
len(data['Book.Title'].unique())
```

```python
#user item interaction  matrix
data2=data.pivot_table(index='User.ID',
                       columns='Book.Title',
                       values='Book.Rating').reset_index(drop=True)
data2
```

```python
# changing user id to original
data2.index=data['User.ID'].unique()
data2
```

```python
# Impute those NaNs with 0 values
data2.fillna(0,inplace=True)
data2
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
```

```python
user_sim = 1 - pairwise_distances( data2.values,metric='cosine')
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
# Set the index and column names to user ids
user_sim_df.index=data['User.ID'].unique()
user_sim_df.columns=data['User.ID'].unique()
user_sim_df
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
# diagnol value with 0
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
# most  similar users
user_sim_df.idxmax(axis=1)[:15]
```
